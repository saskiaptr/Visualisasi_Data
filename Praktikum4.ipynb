Basic Bar Chart dengan Matplotlib

```python
import matplotlib.pyplot as plt

#Data
Jurusan = (
    'Teknik Informatika',
    'Sistem Informasi',
    'Teknik Komputer',
    'Data Science',
)

Jumlah_Mahasiswa = [102, 150, 131, 120]

#Membuat basic bar chart
plt.bar(Jurusan, Jumlah_Mahasiswa, color=['pink', 'red', 'yellow', 'purple'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Jurusan')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Kustomisasi Basic Bar Chart

```python
#Data
Jurusan = (
    'Teknik Informatika',
    'Sistem Informasi',
    'Teknik Komputer',
    'Data Science',
)

Jumlah_Mahasiswa = [102, 150, 131, 120]

#Menambahkan kustomisasi
plt.bar(Jurusan, Jumlah_Mahasiswa, color=['pink', 'red', 'yellow', 'purple'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Jurusan')
plt.xticks(rotation=45)
for i, v in enumerate(Jumlah_Mahasiswa):
  plt.text(i, v + 5, str(v), ha='center') #tambahkan label pada batang
plt.show()
```

Menambahkan Multiple Bar Chart

```python
#Data
Jurusan = (
    'Teknik Informatika',
    'Sistem Informasi',
    'Teknik Komputer',
    'Data Science',
)

Jumlah_Mahasiswa_2024 = [102, 150, 131, 120]

x = range(len(Jurusan))
width = 0.4

plt.bar(x, Jumlah_Mahasiswa, width=width, label='2023', color='pink')
plt.bar([p + width for p in x], Jumlah_Mahasiswa_2024, width=width, label='2024', color='purple')

#Menambahkan kustomisasi
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa per Jurusan (2023 vs 2024)')
plt.xticks(rotation=45)
plt.xticks([p + width / 2 for p in x], Jurusan)
plt.legend()
plt.show()
```

Basic Bar Chart untuk Melihat Pola dan Tren

```python
Jurusan = ['Teknik Informatika', 'Sistem Informasi', 'Teknik Komputer', 'Data Science']
tahun = ['2019', '2020', '2021', '2022', '2023']
Data = {
    'Teknik Informatika': [100, 110, 120, 130, 140],
    'Sistem Informasi': [120, 130, 140, 150, 160],
    'Teknik Komputer': [90, 95, 100, 105, 110],
    'Data Science' :[70, 75, 80, 85, 90]
}

width = 0.2
x = range(len(tahun))

plt.bar(x, Data['Teknik Komputer'], width=width, label='Teknik Komputer', color='pink')
plt.bar([p + width for p in x], Data['Sistem Informasi'], width=width, label='Sistem Informasi', color='red')
plt.bar([p + width * 2 for p in x], Data['Teknik Informatika'], width=width, label='Teknik Informatika', color='')
plt.bar([p + width * 3 for p in x], Data['Data Science'], width=width, label='Data Science', color='purple')

# Memberikan detail pada grafik
plt.title('Tren Jumlah Mahasiswa per Jurusan (2019-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks([p + width * 1.5 for p in x], tahun)
plt.legend()

# Menampilkan grafik
plt.show()
```
