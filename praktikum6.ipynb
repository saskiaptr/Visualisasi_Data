Import Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

Dataset

```python
# dataset
stores = ['store A', 'store B', 'store C']
male = [150, 200, 180]
female = [120, 230, 170]

# data quarter
q1_male = [150, 180, 160]
q1_female = [140, 200, 180]
q2_male = [170, 190, 175]
q2_female = [130, 210, 160]
q3_male = [160, 178, 238]
q3_female = [120, 150, 180]
q4_male = [130, 140, 150]
q4_female = [110, 120, 130]
```

Basic Stacked Vertical Bar Chart

```python
x = np.arange(len(stores))

plt.bar(x, male, label='Male', color='blue')
plt.bar(x, female, bottom=male, label='Female', color='pink')

plt.title('Population by Gender and Store')
plt.xlabel('Stores')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend()

plt.show()
```

Kustomisasi Stacked Vertical Bar Chart

```python
x = np.arange(len(stores))
plt.bar(x, male, label='Male', color='blue')
plt.bar(x, female, bottom=male, label='female', color='green')

plt.title('Sales Transaction by store')
plt.xlabel('Stores')
plt.ylabel('sales')
plt.xticks(x, stores) # Corrected line
plt.legend()

for i in range(len(x)):
    plt.text(x[i], male[i]/2, str(male[i]), ha='center', color='white')
    plt.text(x[i], male[i] + female[i]/2, str(female[i]), ha='center', color='black') # Adjusted y-position for female text
plt.show() # Added plt.show() to display the plot
```

Multiple Stacked Vertical Bar Chart

```python
x = np.arange(len(stores))

# Lebar batang
bar_width = 0.2

# Multiple Stacked Bar Chart
plt.bar(x - bar_width * 1.5, q1_male, label='Q1 Male', color='lightblue', width=bar_width)
plt.bar(x - bar_width * 1.5, q1_female, bottom=q1_male, label='Q1 Female', color='pink', width=bar_width)

plt.bar(x - bar_width / 2, q2_male, label='Q2 Male', color='blue', width=bar_width)
plt.bar(x - bar_width / 2, q2_female, bottom=q2_male, label='Q2 Female', color='red', width=bar_width)

plt.bar(x + bar_width / 2, q3_male, label='Q3 Male', color='green', width=bar_width)
plt.bar(x + bar_width / 2, q3_female, bottom=q3_male, label='Q3 Female', color='yellow', width=bar_width)

plt.bar(x + bar_width * 1.5, q4_male, label='Q4 Male', color='purple', width=bar_width)
plt.bar(x + bar_width * 1.5, q4_female, bottom=q4_male, label='Q4 Female', color='orange', width=bar_width)

# Penyesuaian
plt.xlabel('Stores')
plt.ylabel('Population')
plt.title('Population by Gender and Store (Quarterly)')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Legenda diletakkan di luar grafik

plt.tight_layout()
plt.show()
```
